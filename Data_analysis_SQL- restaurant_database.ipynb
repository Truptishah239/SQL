```python
#Calculate Total Cost
Query = '''SELECT d.Name, SUM(d.Price)
FROM OrdersDishes od
LEFT JOIN Dishes d 
ON od.DishID = d.DishID
WHERE od.OrderID == 100;'''
df = pd.read_sql_query(Query, connection)
df.head()
```

```python
Query = '''SELECT *
FROM Customers c
JOIN Dishes d
ON c.FavoriteDish = d.DishID
WHERE FirstName = 'Cleo' AND LastName = 'Goldwater';'''
df = pd.read_sql_query(Query, connection)
df.head()
```

```python
#Report of Top 5 Customers   
Query = '''SELECT FirstName, LastName, Email, COUNT(OrderID) AS NumberOfOrders
FROM Customers c
JOIN Orders o
ON c.CustomerID = o.CustomerID
GROUP BY c.CustomerID
ORDER BY 4 DESC
LIMIT 5;'''
df = pd.read_sql_query(Query, connection)
df.head()

```
